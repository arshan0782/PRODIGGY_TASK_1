```python
df.drop(['Series Name','Country Name'],axis=1,inplace=True)
```

# Extraction of top 5 countries with respect to total population

```python
# data for total populations 
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']

# SORT data based on the total populatoin for 2022
total_populatoin_sorted = total_population_data.sort_values(by="2022", ascending=False)

# Get the top 5 countries with the highst populztion for 2022
total_top_five_countries = total_populatoin_sorted.head(5)

print("Top five country of total population\n")
print(total_top_five_countries[['Country Code']])
```

# Bar Plot

```python
# Create a bar plot 
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code",data=total_top_five_countries,palette="coolwarm")
plt.title("Top 5 country",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
# Create a bar plot 
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code",data=total_top_five_countries,palette="coolwarm")
plt.title("Top 5 country",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

 # Extraction of top 5 highest countries male populations

```python
#  filtervdata for male populations 
male_population_data = df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']

# SORT data based on the female populatoin for 2022
male_populatoin_sorted = male_population_data.sort_values(by="2022", ascending=False)

# Get the top 5 countries with the highst male population for 2022
male_top_five_countries = male_populatoin_sorted.head(5)

print("Top five country of male population\n")
print(male_top_five_countries[['Country Code']])
```

# Extraction of top 5 highest countries female populations

```python
#  filter data for female populations 
female_population_data = df[df['Series Code'] == 'SP.POP.TOTL.FE.IN']

# SORT data based on the female populatoin for 2022
female_populatoin_sorted = female_population_data.sort_values(by="2022", ascending=False)

# Get the top 5 countries with the highst female population for 2022
female_top_five_countries = female_populatoin_sorted.head(5)

print("Top five country of female population\n")
print(female_top_five_countries[['Country Code']])
```

# Top 5 country highest male and female population in 2022

```python
# Create a bar plot 
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code",data=male_top_five_countries,palette="viridis")
plt.title("Top 5 five countries of male population(2022)",fontsize=10)
plt.xlabel("Male population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
# Create a bar plot 
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code",data=female_top_five_countries,palette="viridis")
plt.title("Top 5 five countries of female population(2022)",fontsize=10)
plt.xlabel("Female population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# STACKED BAR PLOT

# Top 5 countries with male and female  population

```python
# merge male and female population data on country code
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code" ,suffixes=("_male","_female"))
```

```python
# calculate total populations for evry country male or female
merge_data["Total_population"]= merge_data["2022_male"]+ merge_data["2022_female"]
```

```python
merge_data.head()
```

```python
sorted_data = merge_data.sort_values(by="Total_population", ascending=False)
```

# Select top 5 country with high total population

```python
top_5_countries= sorted_data.head(5)
```

```python
#  Create a Stacked bar plpt
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",data= top_5_countries,color="blue",label="female population")
sns.barplot(x="Country Code",y="2022_male",data= top_5_countries,bottom=top_5_countries["2022_female"],color="green" , label="Male Population")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

# Bottom 5 country with male and female total population (2022)

```python
#Select top 5 country with high total population
Bottom_5_countries = sorted_data.tail(5)
```

```python
#  Create a Stacked bar plpt
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",data= Bottom_5_countries,color="blue",label="female population")
sns.barplot(x="Country Code",y="2022_male",data= Bottom_5_countries,bottom=top_5_countries["2022_female"],color="green" , label="Male Population")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```
